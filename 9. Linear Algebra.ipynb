#### Simple Array Operations

```python
import numpy as np
```

```python
a = np.array([[1.0, 2.0], [3.0, 4.0]])
a
```

```python
a.transpose()
```

```python
np.linalg.inv(a)
```

```python
u = np.eye(2) # unit 2x2 matrix; "eye" represents "I"
u
```

```python
j = np.array([[0.0, -1.0], [1.0, 0.0]])
j @ j        # matrix product
```

```python
np.trace(u)  # trace
```

```python
y = np.array([[5.], [7.]])
np.linalg.solve(a, y)
```

```python
np.linalg.eig(j)
```
